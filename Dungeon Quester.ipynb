```python
import random

# Step 1: Player Setup
player_name = input("Enter your adventurer's name: ")
health = 10
inventory = []

print(f"Welcome, {player_name}! Your quest begins now.")

# Step 2: Treasure Dictionary
treasures = {
    "Golden Chalice": 50,
    "Jeweled Dagger": 40,
    "Ancient Coin": 30,
    "Mystic Ring": 25,
    "Emerald Idol": 60
}

# Step 3: Game Loop - Moving through 5 rooms
for room in range(1, 6):
    print(f"\nüè∞ You are in room {room}")
    
    while True:
        print("\nChoose an action:")
        print("1. Search for treasure")
        print("2. Move to next room")
        print("3. Check health and inventory")
        print("4. Quit the game")
        choice = input("Enter your choice (1-4): ")

        # Step 4: Treasure or Trap
        if choice == "1":
            found = random.choice(["treasure", "trap"])
            if found == "treasure":
                item = random.choice(list(treasures.keys()))
                inventory.append(item)
                print(f"You found a {item}! It's worth {treasures[item]} gold.")
            else:
                health -= 2
                print("üí• Oh no! You triggered a trap and lost 2 health.")
            break  # After search, move to next room
        
        elif choice == "2":
            print("You move carefully to the next room...")
            break  # Move to next room
        
        elif choice == "3":
            print(f"‚ù§Ô∏è Health: {health}")
            print(f"üéí Inventory: {inventory if inventory else 'Empty'}")
        
        elif choice == "4":
            print("You‚Äôve chosen to end your quest.")
            room = 5  # Force game end
            break
        
        else:
            print("Invalid choice. Please try again.")
    
    # Step 5: Health Check
    if health <= 0:
        print("\nüíÄ You‚Äôve succumbed to your wounds. Game Over!")
        break

# Step 6: End of Game Summary
print("\nüèÅ Your adventure has ended.")
total_value = sum(treasures[item] for item in inventory)
print(f"‚ù§Ô∏è Final Health: {health}")
print(f"üéí Final Inventory: {inventory if inventory else 'Empty'}")
print(f"üí∞ Total Treasure Value: {total_value} gold")
```
